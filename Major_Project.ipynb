# ***Forcasting Model***

```
df
```

```
df.fillna(method = "bfill")
```

```
df.rename(columns = {'new_cases': 'y', 'date': 'ds'},inplace = True)
```

```
df
```

```
df_new = df[['ds','y']]
df_new
```

```
df_new.drop(df.tail(1).index,inplace = True)
```

```
df_new
```

```
df_new.info()
```

```
import matplotlib.pyplot as plt
plt.figure(figsize = (15,5))
plt.scatter(df_new['ds'],df_new['y'])
plt.show()
```

```
import fbprophet
op = fbprophet.Prophet()
op.fit(df_new)
```

```
forecast = op.make_future_dataframe(periods = 365,freq = 'D')
forecast = op.predict(forecast)
```

```
forecast.columns
```

```
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
```

```
op.plot(forecast,xlabel = 'date',ylabel ='new_cases');
```
